```python
!pip install pymupdf
import fitz
import json
import re

# from google.colab import files
# uploaded=files.upload()
output=[]
text=""
doc=fitz.open('LenderFeesWorksheetNew.pdf')
for page in doc:
  text+= page.get_text("text", sort=True)

Applicant=re.search(r"Applicants?:\s*(.+?)\s+Application\s+No:",text)
loanProgram=re.search(r"Loan\s+Program:\s*(.+)",text)
loanAmount=re.search(r"Total\s+Loan\s+Amount:\s*\$?\s*([\d,]+(?:\.\d{2})?)",text)
intRate=re.search(r"Interest\s+Rate:\s*(([\d.]+)\s*%?)",text)
term=re.search(r"Term/Due\s+In:\s*(.+)",text)
pI=re.search(r"Principal\s+\&\s+Interest\s*(.+)", text)
closing=re.search(r"Est\.\s+Closing\s+Costs\s*(.+?)\s+Homeowner\s+Assn\.", text)
totPay=re.search(r"Total\s+Monthly\s+Payment\s*(.+)",text)


# print("Applicant:", Applicant.group(1).strip() if Applicant else None)
# print("Loan Program:", loanProgram.group(1).strip() if loanProgram else None)
# print("Loan Amount:", loanAmount.group(1).strip() if loanAmount else None)
# print("Interest Rate:", intRate.group(1).strip() if intRate else None)
# print("Term/Due In:", term.group(1).strip() if term else None)
# print("Principal & Interest:", pI.group(1).strip() if pI else None)
print("Est. Closing Costs:", closing.group(1).strip() if closing else None)
print("Total Monthly Payment:", totPay.group(1).strip() if totPay else None)

def findBbox(page, t):

    rect=page.search_for(t)
    if not rect:
      return None
    # x0 = min(r.x0 for r in rect)
    # y0 = min(r.y0 for r in rect)
    # x1 = max(r.x1 for r in rect)
    # y1 = max(r.y1 for r in rect)
    # return[x0,y0,x1,y1]
    return(rect[0].x0,rect[0].y0,rect[0].x1,rect[0].y1)


output.append({Applicant.group(1).strip():findBbox(doc[0], Applicant.group(1).strip()),
               loanProgram.group(1).strip():findBbox(doc[0], loanProgram.group(1).strip()),
               loanAmount.group(1).strip():findBbox(doc[0], loanAmount.group(1).strip()),
               intRate.group(1).strip():findBbox(doc[0], intRate.group(1).strip()),
               term.group(1).strip():findBbox(doc[0], term.group(1).strip()),
               pI.group(1).strip():findBbox(doc[0], pI.group(1).strip()),
               closing.group(1).strip():findBbox(doc[0], closing.group(1).strip()),
               totPay.group(1).strip():findBbox(doc[0], totPay.group(1).strip()),



})

print (output)
```
