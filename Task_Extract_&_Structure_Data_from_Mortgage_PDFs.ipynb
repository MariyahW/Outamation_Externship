```python
!pip install pymupdf
import requests
import fitz

# url='https://raw.githubusercontent.com/MariyahW/Outamation_Externship/refs/heads/main/LenderFeesWorksheetNew.pdf'
# res= requests.get(url)

# from google.colab import files
# uploaded=files.upload()

# Load uploaded pdf
doc=fitz.open('LenderFeesWorksheetNew.pdf')

print(f"Total Pages: { doc.page_count}" )

print("PDF Metadata:")
doc.metadata


# Loop through all pages and extract text
for page_num in range(len(doc)):
  page=doc[page_num]
  text=page.get_text("blocks", sort=True)
  print(f" --- Page {page_num+1} ---")
  for block in text:
    print(block,"\n")

```
