```python
!pip install pymupdf
import requests
import fitz
import json

# url='https://raw.githubusercontent.com/MariyahW/Outamation_Externship/refs/heads/main/LenderFeesWorksheetNew.pdf'
# res= requests.get(url)

# from google.colab import files
# uploaded=files.upload()

# Load uploaded pdf
doc=fitz.open('LenderFeesWorksheetNew.pdf')

print(f"Total Pages: { doc.page_count}" )

print("PDF Metadata:")
doc.metadata

all_pages_data=[]

# Loop through all pages and extract text
for page_num in range(len(doc)):
  page=doc[page_num]
  words=page.get_text("words", sort=True)
  page_data=[]
  for w in words:
    x0,y0,x1,y1,text, block_no, line_no, word_no=w
    entry = {
        "text":text,
        "bbox":[x0,y0,x1,y1],
        "page":page_num+1
    }
    page_data.append(entry)

  all_pages_data.append({
        "page":page_num+1,
        "content": page_data
    })
json_output=json.dumps(all_pages_data, indent=2)

with open("loan_sheet_extraction.json", "w") as f:
   f.write(json_output)

print("JSON file saved as loan_worksheet_extraction.json")
print(json_output)

```
