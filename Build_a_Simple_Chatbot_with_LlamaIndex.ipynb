```python
!pip -q install -U google-generativeai llama-index llama-index-llms-gemini

import os
import google.generativeai as genai

# ---------------------------
# 1) API KEY (paste your NEW rotated key here)
# ---------------------------
os.environ["GOOGLE_API_KEY"] = "AIzaSyDMoxzXH55VfZv-IP0GjvX8IlmtSeqoDU8"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# ---------------------------
# 2) Discover available models (no guessing)
# ---------------------------
valid_models = []
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        valid_models.append(m.name)

if not valid_models:
    raise RuntimeError("No models available for this API key. Check your key/project access.")

print("Valid models:")
for m in valid_models:
    print(" -", m)

# Pick the first available model (you can change index if you want a different one)
MODEL_NAME = valid_models[0]
print("\nâœ… Using model:", MODEL_NAME)

# ---------------------------
# 3) LlamaIndex Gemini init (uses the discovered model)
# ---------------------------
from llama_index.llms.gemini import Gemini
from llama_index.core.llms import ChatMessage

llm = Gemini(model=MODEL_NAME)

# ---------------------------
# 4) Simple interactive chatbot
# ---------------------------
def simple_chatbot():
    print("\nðŸ¦™ Simple Gemini Chatbot (LlamaIndex) ðŸ¦™")
    print("Type 'exit' to end the conversation")
    print("-" * 50)

    messages = []

    while True:
        user_input = input("\nYou: ").strip()

        if user_input.lower() in ["exit", "quit", "bye"]:
            print("\nChatbot: Goodbye! Have a great day!")
            break

        messages.append(ChatMessage(role="user", content=user_input))

        try:
            response = llm.chat(messages)
            answer = response.message.content
            print(f"\nChatbot: {answer}")
            messages.append(ChatMessage(role="assistant", content=answer))

        except Exception as e:
            print(f"\nError: {e}")
            print("If this persists, restart runtime and rerun this cell.")

# Run chatbot
simple_chatbot()
```
