```python
!apt-get install -y tesseract-ocr
!pip install pytesseract easyocr paddleocr opencv-python pillow pymupdf

import fitz
from PIL import Image
import pytesseract
import numpy as np
import easyocr
import cv2



doc = fitz.open("mortgage.pdf")
images = []

for page in doc:
    pix = page.get_pixmap(dpi=300)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    images.append(img)

# --- Tesseract ---
tesseract_text = []
for img in images:
    text = pytesseract.image_to_string(img)
    tesseract_text.append(text)

print(tesseract_text[0][:1500])

# --- EasyOCR ---
reader = easyocr.Reader(['en'])
easyocr_results = []

for img in images:
    img_np = np.array(img)
    result = reader.readtext(img_np)
    easyocr_results.append(result)

print("EasyOCR completed successfully")

# --- PaddleOCR (init once per runtime) ---
try:
    paddle
    print("PaddleOCR already initialized — reusing it.")
except NameError:
    from paddleocr import PaddleOCR
    paddle = PaddleOCR(use_angle_cls=True, lang='en')
    print("PaddleOCR initialized.")

# --- Draw EasyOCR boxes (optional) ---
img_cv = cv2.cvtColor(np.array(images[0]), cv2.COLOR_RGB2BGR)
for box, text, conf in easyocr_results[0]:
    pts = np.array(box).astype(int)
    cv2.polylines(img_cv, [pts], True, (0,255,0), 2)

# --- Paddle result page 0 ---
paddle_result_page0 = paddle.ocr(np.array(images[0]), cls=True)
paddle_result_page0[:1]
```
