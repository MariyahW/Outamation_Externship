```python

# import regex, requests, nltk, contractions, spell checker
!pip install contractions
!pip install pyspellchecker
from spellchecker import SpellChecker
import contractions
import nltk
from nltk.corpus import stopwords
import requests
#import string
import re

spell=SpellChecker()

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def speller(text):
  words=text.split()
  fixed=[spell.correction(word) for word in words]
  return " ".join(fixed)

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)

#bring in file that needs to be cleaned
url='https://raw.githubusercontent.com/MariyahW/Outamation_Externship/refs/heads/main/text_sample.txt'

# function to clean text
# this function first makes everything lowercase
# it will then remove everything that is not a number, letter, or a space
# then it will remove extra spacing, contractions, and stopwords, lastly spell check
def cleanText(text):
  text=text.lower()
  #trans=str.maketrans("","",string.punctuation)
  #text=text.translate(trans)
  text=re.sub(r"[^a-z0-9\s]","",text)
  text=re.sub(r"\s+"," ",text)
  text=contractions.fix(text)
  text=remove_stopwords(text)
  text=speller(text)
  return text

response= requests.get(url)
messy=response.text

text=cleanText(messy)

print(text)
```
